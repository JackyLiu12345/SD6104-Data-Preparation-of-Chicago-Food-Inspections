```python
"""
Food Inspection Data - Exploratory Data Analysis (EDA) Script
Author: [Liu Tingxu]
Description: Comprehensive exploratory data analysis for Chicago Food Inspection dataset
Requirements: pandas, numpy, matplotlib, seaborn
"""

import pandas as pd
import numpy as np
from datetime import datetime
import warnings
warnings.filterwarnings('ignore')

def load_data(file_path):
    """Load the dataset from CSV file"""
    print("=" * 80)
    print("FOOD INSPECTION DATA EXPLORATORY ANALYSIS")
    print("=" * 80)
    print(f"Loading data from: {file_path}")
    
    try:
        df = pd.read_csv(file_path, low_memory=False)
        print(f"✓ Data loaded successfully: {df.shape[0]:,} rows × {df.shape[1]:,} columns")
        return df
    except FileNotFoundError:
        print(f"✗ Error: File not found at '{file_path}'")
        exit()
    except Exception as e:
        print(f"✗ Error loading data: {e}")
        exit()

def print_data_overview(df):
    """Print basic overview of the dataset"""
    print("\n" + "=" * 80)
    print("1. DATASET OVERVIEW")
    print("=" * 80)
    
    num_rows, num_cols = df.shape
    print(f"Dataset Dimensions: {num_rows:,} rows × {num_cols:,} columns")
    
    return num_rows, num_cols

def print_column_summary(df):
    """Print summary of all columns and their data types"""
    print("\n" + "-" * 80)
    print("2. COLUMN SUMMARY")
    print("-" * 80)
    
    # Create a formatted table of columns
    print(f"{'No.':<4} {'Column Name':<25} {'Data Type':<15} {'Unique Values':<15} {'Null Values':<12} {'Null %':<8}")
    print("-" * 80)
    
    for i, col in enumerate(df.columns, 1):
        dtype = str(df[col].dtype)
        unique_count = df[col].nunique()
        null_count = df[col].isnull().sum()
        null_pct = (null_count / len(df)) * 100
        
        print(f"{i:<4} {col:<25} {dtype:<15} {unique_count:<15,} {null_count:<12,} {null_pct:<8.2f}")

def analyze_single_columns(df, num_rows):
    """Perform detailed analysis of each column"""
    print("\n" + "-" * 80)
    print("3. DETAILED COLUMN ANALYSIS")
    print("-" * 80)
    
    analysis_results = []
    
    for column in df.columns:
        col_info = {
            'Column': column,
            'Data_Type': str(df[column].dtype),
            'Unique_Values': df[column].nunique(),
            'Null_Values': df[column].isnull().sum(),
            'Null_Percentage': round((df[column].isnull().sum() / num_rows) * 100, 2)
        }
        
        # Check if column is numeric
        is_numeric = pd.api.types.is_numeric_dtype(df[column])
        
        # Handle top 5 values and their counts
        if df[column].nunique() <= 20:
            # 20 or fewer unique values, show all values
            value_counts = df[column].value_counts(dropna=False)
            col_info['Value_Distribution'] = format_value_counts(value_counts)
        else:
            # More than 20 unique values
            if is_numeric:
                # High cardinality numeric column, show statistics
                if df[column].notna().any():
                    stats = f"Range: [{df[column].min():.2f}, {df[column].max():.2f}], "
                    stats += f"Mean: {df[column].mean():.2f}, "
                    stats += f"Std: {df[column].std():.2f}"
                    col_info['Value_Distribution'] = stats
                else:
                    col_info['Value_Distribution'] = 'All Null Values'
            else:
                # High cardinality non-numeric column, show top 5
                top_vals = df[column].value_counts(dropna=False).head(5)
                col_info['Value_Distribution'] = format_value_counts(top_vals)
        
        analysis_results.append(col_info)
    
    # Print analysis results in a clean format
    print(f"{'Column':<25} {'Type':<10} {'Unique':<8} {'Null':<6} {'Null%':<7} {'Top Values / Statistics'}")
    print("-" * 80)
    
    for result in analysis_results:
        col_name = result['Column']
        if len(col_name) > 24:
            col_name = col_name[:21] + "..."
        
        # Print main row
        print(f"{col_name:<25} {result['Data_Type']:<10} {result['Unique_Values']:<8,} "
              f"{result['Null_Values']:<6,} {result['Null_Percentage']:<6.2f} ", end="")
        
        # Print value distribution (first line)
        dist_lines = str(result['Value_Distribution']).split('\n')
        if dist_lines:
            print(dist_lines[0][:50] + ("..." if len(dist_lines[0]) > 50 else ""))
        
        # Print additional lines if needed
        for line in dist_lines[1:]:
            print(" " * 56 + line[:50] + ("..." if len(line) > 50 else ""))

def format_value_counts(value_counts, max_items=5):
    """Format value counts for display"""
    if len(value_counts) == 0:
        return "No data"
    
    result = []
    for i, (value, count) in enumerate(value_counts.items()):
        if i >= max_items:
            break
        
        # Handle NaN values
        if pd.isna(value):
            value_str = "NaN"
        else:
            value_str = str(value)
            
        # Truncate long values
        if len(value_str) > 30:
            value_str = value_str[:27] + "..."
        
        result.append(f"{value_str}: {count:,}")
    
    return "\n".join(result)
    

def main():
    """Main function to run the EDA analysis"""
    # Set the file path
    file_path = r"E:\下载\Food_Inspections_20260327.csv"  # Update with your file path
    
    # Load data
    df = load_data(file_path)
    
    # Perform analysis steps
    num_rows, num_cols = print_data_overview(df)
    print_column_summary(df)
    analyze_single_columns(df, num_rows)

if __name__ == "__main__":
    main()
```

```python
import re

# Load data
file_path = r"E:\下载\Food_Inspections_20260327.csv"
df = pd.read_csv(file_path, low_memory=False)

print("=" * 60)
print("Starting Data Cleaning")
print("=" * 60)
print(f"Original data shape: {df.shape[0]} rows × {df.shape[1]} columns")

# 1. Process Facility Type column
print("\n1. Processing Facility Type Column")
print("-" * 40)
# Replace null values with "Others"
facility_null_count = df['Facility Type'].isnull().sum()
df['Facility Type'] = df['Facility Type'].fillna('Others')
print(f"  Replaced {facility_null_count} null values with 'Others'")
print(f"  'Others' now has {df['Facility Type'].value_counts().get('Others', 0)} records")

# 2. Delete Location, City, State columns
print("\n2. Deleting Location, City, State Columns")
print("-" * 40)
cols_to_drop = ['Location', 'City', 'State']
cols_exist = [col for col in cols_to_drop if col in df.columns]
df = df.drop(columns=cols_exist)
print(f"  Deleted columns: {cols_exist}")
print(f"  Data shape after deletion: {df.shape[0]} rows × {df.shape[1]} columns")

# 3. Process Zip column
print("\n3. Processing Zip Column")
print("-" * 40)
# Record original row count
original_rows = df.shape[0]

# Delete null values
zip_null_count = df['Zip'].isnull().sum()
df = df.dropna(subset=['Zip'])
print(f"  a) Deleted {zip_null_count} null values")

# Delete rows outside specified zip code range
# First convert Zip column to numeric type
df['Zip'] = pd.to_numeric(df['Zip'], errors='coerce')

# Apply new range filter conditions
# Keep: 60600-60699 OR 60290 OR 60701
zip_mask = (
    ((df['Zip'] >= 60600) & (df['Zip'] <= 60699)) |  # 60600-60699 range
    (df['Zip'] == 60290) |                           # equals 60290
    (df['Zip'] == 60701)                             # equals 60701
)
df = df[zip_mask].copy()

# Count deleted rows
rows_after_zip = df.shape[0]
rows_removed_zip = original_rows - rows_after_zip

# Count records in each zip code range
zip_606 = ((df['Zip'] >= 60600) & (df['Zip'] <= 60699)).sum()
zip_60290 = (df['Zip'] == 60290).sum()
zip_60701 = (df['Zip'] == 60701).sum()

print(f"  b) Deleted {rows_removed_zip} rows outside specified zip code range")
print(f"  c) Unique zip code count: {df['Zip'].nunique()}")
print(f"  d) Zip code range: {df['Zip'].min()} to {df['Zip'].max()}")
print(f"  e) Record counts by zip code range:")
print(f"     - 60600-60699: {zip_606:,} records")
print(f"     - 60290: {zip_60290:,} records")
print(f"     - 60701: {zip_60701:,} records")

# 4. Process License # column
print("\n4. Processing License # Column")
print("-" * 40)
original_rows = df.shape[0]

# Delete null values
license_null_count = df['License #'].isnull().sum()
df = df.dropna(subset=['License #'])
print(f"  a) Deleted {license_null_count} null values")

# Delete rows with value 0
# First convert to numeric type
df['License #'] = pd.to_numeric(df['License #'], errors='coerce')
zeros_mask = (df['License #'] == 0)
zero_count = zeros_mask.sum()
df = df[~zeros_mask].copy()

rows_after_license = df.shape[0]
rows_removed_license = original_rows - rows_after_license
print(f"  b) Deleted {zero_count} rows with value 0")
print(f"  c) Total rows deleted: {rows_removed_license}")
print(f"  d) License # range: {df['License #'].min()} to {df['License #'].max()}")

# 5. Process Inspection Type column
print("\n5. Processing Inspection Type Column")
print("-" * 40)
original_rows = df.shape[0]

# Delete null values
inspection_type_null_count = df['Inspection Type'].isnull().sum()
df = df.dropna(subset=['Inspection Type'])

rows_after_inspection = df.shape[0]
rows_removed_inspection = original_rows - rows_after_inspection
print(f"  Deleted {inspection_type_null_count} null values")
print(f"  Total rows deleted: {rows_removed_inspection}")

# 6. Process Risk column
print("\n6. Processing Risk Column")
print("-" * 40)
original_rows = df.shape[0]

# Delete null values
risk_null_count = df['Risk'].isnull().sum()
df = df.dropna(subset=['Risk'])
print(f"  a) Deleted {risk_null_count} null values")

# Delete rows with value "All"
all_mask = (df['Risk'] == 'All')
all_count = all_mask.sum()
df = df[~all_mask].copy()

rows_after_risk = df.shape[0]
rows_removed_risk = original_rows - rows_after_risk
print(f"  b) Deleted {all_count} rows with value 'All'")
print(f"  c) Total rows deleted: {rows_removed_risk}")
print(f"  d) Risk unique values: {df['Risk'].unique().tolist()}")

# 7. Process Inspection Date column
print("\n7. Processing Inspection Date Column")
print("-" * 40)
# Convert to datetime format
df['Inspection Date'] = pd.to_datetime(df['Inspection Date'], format='%m/%d/%Y', errors='coerce')

# Check conversion failures
conversion_failures = df['Inspection Date'].isnull().sum()
if conversion_failures > 0:
    print(f"  Warning: {conversion_failures} date conversion failures")

# Extract year, month, day
df['Inspection Year'] = df['Inspection Date'].dt.year
df['Inspection Month'] = df['Inspection Date'].dt.month
df['Inspection Day'] = df['Inspection Date'].dt.day

# Delete original column
df = df.drop(columns=['Inspection Date'])

print(f"  Added three columns: Inspection Year, Inspection Month, Inspection Day")
print(f"  Date range: {df['Inspection Year'].min()} to {df['Inspection Year'].max()}")

# 8. Process Violations column
print("\n8. Processing Violations Column")
print("-" * 40)

def extract_violation_terms(violation_text):
    """
    Extract clause numbers from violation description
    Improved extraction function that correctly handles multiple violation clauses
    
    Args:
        violation_text (str): Original violation description text
        
    Returns:
        str: Extracted clause numbers, comma-separated, e.g., "23,37,38"
        np.nan: If no clause numbers are extracted
    """
    if pd.isna(violation_text):
        return np.nan
    
    text = str(violation_text)
    
    # 1. First split multiple violation clauses by "|"
    clauses = [clause.strip() for clause in text.split('|')]
    
    all_matches = []
    
    for clause in clauses:
        # 2. For each clause, separate main description and Comments part
        main_part = clause
        
        # Comments markers can have various forms
        comments_markers = [" - Comments:", "Comments:", " - ", "COMMENTS:"]
        for marker in comments_markers:
            if marker in main_part:
                parts = main_part.split(marker, 1)
                if len(parts) > 1:
                    main_part = parts[0]  # Keep only the main description part
                    break
        
        # 3. Extract clause numbers from the main description part
        # Pattern: Digits at the beginning, followed by a dot, then a space
        # Need to ensure this is a clause number, not a number in Comments
        # Clause numbers are usually at the beginning of a clause
        pattern = r'^\s*(\d+)\.\s'
        match = re.search(pattern, main_part)
        if match:
            # Extract clause number
            term = match.group(1)
            try:
                # Remove leading zeros
                term_clean = str(int(term))
                all_matches.append(term_clean)
            except ValueError:
                all_matches.append(term)
        else:
            # Backup pattern: Try to match other clause numbers in the main description
            backup_pattern = r'\b(\d+)\.\s+[A-Z]'
            backup_matches = re.findall(backup_pattern, main_part)
            for term in backup_matches:
                try:
                    term_clean = str(int(term))
                    all_matches.append(term_clean)
                except ValueError:
                    all_matches.append(term)
    
    if all_matches:
        # Remove duplicates and sort
        unique_matches = sorted(set(all_matches), key=lambda x: int(x) if x.isdigit() else 0)
        return ','.join(unique_matches)
    
    return np.nan

# Apply extraction function
df['Violation Terms'] = df['Violations'].apply(extract_violation_terms)

# Count extraction results
extracted_count = df['Violation Terms'].notna().sum()
no_violation_count = df['Violations'].isna().sum()
total_rows = df.shape[0]

print(f"  a) Successfully extracted clause numbers from {extracted_count} records")
print(f"  b) {no_violation_count} records have no violations")
print(f"  c) Example extraction results:")


'''
# You can choose whether generate output file
# 9. Save cleaned data
print("\n9. Saving Cleaning Results")
print("-" * 40)
# Generate output filename
output_file = 'Food_Inspections_Cleaned.csv'

# Save data
df.to_csv(output_file, index=False)

print(f"  Cleaned data saved to: {output_file}")
'''

# Show cleaned data overview
print("\n" + "=" * 60)
print("Cleaning Complete! Data Overview")
print("=" * 60)
print(f"Total records: {df.shape[0]:,}")
print(f"Total columns: {df.shape[1]}")
print(f"Data type distribution:")
for dtype, count in df.dtypes.value_counts().items():
    print(f"  {dtype}: {count} columns")

print(f"\nColumn information:")
for i, col in enumerate(df.columns, 1):
    null_count = df[col].isnull().sum()
    null_pct = null_count / len(df) * 100
    unique_count = df[col].nunique()
    print(f"  {i:2d}. {col:25s} Null: {null_count:6,d} ({null_pct:5.2f}%) Unique: {unique_count:6,d}")

print("\n" + "=" * 60)
print("Data Cleaning Process Complete!")
print("=" * 60)
```

```python
import pandas as pd
import numpy as np
from itertools import combinations
from collections import defaultdict
from typing import Set, Tuple, List, Dict
import warnings
warnings.filterwarnings('ignore')

class FullINDDiscovery:
    """
    Full multi-column Inclusion Dependency Discovery (with confidence) - Using all data
    Discovers inclusion dependencies between all columns
    """
    
    def __init__(self, df: pd.DataFrame, min_confidence: float = 0.95, max_level: int = 2):
        """
        Initialize
        
        Args:
            df: DataFrame (all data)
            min_confidence: Minimum confidence threshold (0-1), dependencies below this are not output
            max_level: Maximum combination level (default 2, as 3+ columns cause combinatorial explosion)
        """
        self.df = df
        self.min_confidence = min_confidence
        self.max_level = max_level
        self.cols = list(df.columns)
        
        print(f"Initialization complete:")
        print(f"  Number of rows: {len(df):,}")
        print(f"  Number of columns: {len(self.cols)}")
        print(f"  Minimum confidence: {min_confidence}")
        print(f"  Maximum combination level: {max_level}")
        print("-" * 70)
        
        # Precompute value sets for each column (using all data)
        print("Precomputing value sets for each column...")
        self.col_values = {}
        for i, col in enumerate(self.cols):
            self.col_values[col] = set(self.df[col].dropna().astype(str).unique())
            if (i + 1) % 10 == 0:
                print(f"  Completed {i+1}/{len(self.cols)} columns")
        print(f"  Complete! Total {len(self.cols)} columns")
        print("-" * 70)
    
    def get_combined_values(self, cols: Tuple[str]) -> Set[Tuple]:
        """
        Get combined value set for multiple columns (using all data)
        
        Args:
            cols: Tuple of column names
            
        Returns:
            Set of combined values, each value is a tuple
        """
        if len(cols) == 1:
            return self.col_values[cols[0]]
        
        values = set()
        for _, row in self.df.iterrows():
            key = []
            valid = True
            for col in cols:
                val = row[col]
                if pd.isna(val):
                    valid = False
                    break
                key.append(str(val))
            
            if valid:
                values.add(tuple(key))
        
        return values
    
    def check_ind(self, dep_cols: Tuple[str], ref_cols: Tuple[str]) -> Dict:
        """
        Check inclusion dependency and compute confidence
        
        Args:
            dep_cols: Dependent column tuple
            ref_cols: Referenced column tuple
            
        Returns:
            dict: Contains confidence, match counts, etc.
        """
        # Get value sets
        dep_values = self.get_combined_values(dep_cols)
        ref_values = self.get_combined_values(ref_cols)
        
        dep_count = len(dep_values)
        if dep_count == 0:
            return {
                'confidence': 1.0,
                'is_ind': True,
                'dep_count': 0,
                'ref_count': len(ref_values),
                'matched_count': 0,
                'missing_count': 0
            }
        
        # Calculate matches
        matched = dep_values.intersection(ref_values)
        matched_count = len(matched)
        confidence = matched_count / dep_count
        is_ind = (confidence == 1.0)
        
        return {
            'confidence': confidence,
            'is_ind': is_ind,
            'dep_count': dep_count,
            'ref_count': len(ref_values),
            'matched_count': matched_count,
            'missing_count': dep_count - matched_count
        }
    
    def generate_candidates_1col(self) -> List[Tuple[Tuple, Tuple]]:
        """
        Generate single-column candidates (all column pairs)
        """
        candidates = []
        for dep in self.cols:
            for ref in self.cols:
                if dep != ref:
                    candidates.append(((dep,), (ref,)))
        return candidates
    
    def generate_candidates_2col(self, confirmed_1col: List[Tuple[Tuple, Tuple]]) -> List[Tuple[Tuple, Tuple]]:
        """
        Generate two-column candidates from single-column INDs
        
        Based on Apriori property:
        If (A,B) ⊆ (C,D) holds, then A ⊆ C and B ⊆ D must also hold
        """
        # Build mapping: dep -> possible refs
        dep_to_refs = defaultdict(set)
        for dep_tuple, ref_tuple in confirmed_1col:
            dep_to_refs[dep_tuple[0]].add(ref_tuple[0])
        
        deps = list(dep_to_refs.keys())
        candidates = []
        
        # Generate all possible two-column combinations
        for i in range(len(deps)):
            for j in range(i + 1, len(deps)):
                dep1, dep2 = deps[i], deps[j]
                
                for ref1 in dep_to_refs.get(dep1, set()):
                    for ref2 in dep_to_refs.get(dep2, set()):
                        # Candidate 1: (dep1, dep2) ⊆ (ref1, ref2)
                        candidates.append(((dep1, dep2), (ref1, ref2)))
                        
                        # Candidate 2: (dep1, dep2) ⊆ (ref1, ref1)  # Map to same column
                        if ref1 != ref2:
                            candidates.append(((dep1, dep2), (ref1, ref1)))
                            candidates.append(((dep1, dep2), (ref2, ref2)))
        
        # Remove duplicates
        return list(set(candidates))
    
    def discover_all_inds(self) -> pd.DataFrame:
        """
        Discover all inclusion dependencies (single and multi-column)
        """
        all_results = []
        
        # ========== Level 1: Single Column ==========
        print("\n" + "=" * 70)
        print("Level 1: Single Column Inclusion Dependency Discovery")
        print("=" * 70)
        
        candidates_1col = self.generate_candidates_1col()
        print(f"Generated {len(candidates_1col)} candidates")
        
        confirmed_1col = []
        
        # Process in batches, show progress
        total = len(candidates_1col)
        for idx, (dep_tuple, ref_tuple) in enumerate(candidates_1col):
            if (idx + 1) % 100 == 0:
                print(f"  Progress: {idx+1}/{total} ({100*(idx+1)/total:.1f}%)")
            
            result = self.check_ind(dep_tuple, ref_tuple)
            
            if result['confidence'] >= self.min_confidence:
                confirmed_1col.append((dep_tuple, ref_tuple))
                all_results.append({
                    'level': 1,
                    'dependent': dep_tuple[0],
                    'referenced': ref_tuple[0],
                    'confidence': result['confidence'],
                    'is_ind': result['is_ind'],
                    'dep_count': result['dep_count'],
                    'ref_count': result['ref_count'],
                    'matched_count': result['matched_count']
                })
        
        print(f"Complete! Confirmed {len(confirmed_1col)} (confidence >= {self.min_confidence})")
        
        # Display all single-column results
        if all_results:
            df_results = pd.DataFrame(all_results)
            df_results = df_results.sort_values('confidence', ascending=False)
            
            print(f"\nSingle-column inclusion dependencies (total {len(df_results)}):")
            print("-" * 80)
            for i, row in df_results.iterrows():
                status = "✓" if row['is_ind'] else "~"
                print(f"  {status} {row['dependent']:30s} ⊆ {row['referenced']:30s} "
                      f"conf={row['confidence']:.6f} ({row['matched_count']}/{row['dep_count']})")
        
        # ========== Level 2: Two Columns ==========
        if self.max_level >= 2 and confirmed_1col:
            print("\n" + "=" * 70)
            print("Level 2: Two-Column Inclusion Dependency Discovery")
            print("=" * 70)
            
            candidates_2col = self.generate_candidates_2col(confirmed_1col)
            print(f"Generated {len(candidates_2col)} candidates")
            
            # Process all candidates without limit
            print(f"Validating {len(candidates_2col)} two-column candidates...")
            
            confirmed_2col = []
            total = len(candidates_2col)
            
            for idx, (dep_tuple, ref_tuple) in enumerate(candidates_2col):
                if (idx + 1) % 100 == 0:
                    print(f"  Progress: {idx+1}/{total} ({100*(idx+1)/total:.1f}%)")
                
                result = self.check_ind(dep_tuple, ref_tuple)
                
                if result['confidence'] >= self.min_confidence:
                    confirmed_2col.append((dep_tuple, ref_tuple))
                    dep_str = f"({', '.join(dep_tuple)})"
                    ref_str = f"({', '.join(ref_tuple)})"
                    all_results.append({
                        'level': 2,
                        'dependent': dep_str,
                        'referenced': ref_str,
                        'confidence': result['confidence'],
                        'is_ind': result['is_ind'],
                        'dep_count': result['dep_count'],
                        'ref_count': result['ref_count'],
                        'matched_count': result['matched_count']
                    })
            
            print(f"Complete! Confirmed {len(confirmed_2col)} (confidence >= {self.min_confidence})")
            
            # Display two-column results
            if confirmed_2col:
                print(f"\nTwo-column inclusion dependencies (total {len(confirmed_2col)}):")
                print("-" * 80)
                temp_results = [r for r in all_results if r['level'] == 2]
                temp_df = pd.DataFrame(temp_results).sort_values('confidence', ascending=False)
                for _, row in temp_df.iterrows():
                    status = "✓" if row['is_ind'] else "~"
                    print(f"  {status} {row['dependent']:45s} ⊆ {row['referenced']:45s} "
                          f"conf={row['confidence']:.6f}")
        
        # Convert to DataFrame
        results_df = pd.DataFrame(all_results)
        
        if not results_df.empty:
            results_df = results_df.sort_values(['level', 'confidence'], ascending=[True, False])
        
        return results_df
    
    def print_summary(self, results_df: pd.DataFrame):
        """
        Print analysis summary
        """
        print("\n" + "=" * 80)
        print("Inclusion Dependency Analysis Summary")
        print("=" * 80)
        
        if results_df.empty:
            print("No inclusion dependencies found")
            return
        
        # Statistics by level
        for level in [1, 2]:
            level_df = results_df[results_df['level'] == level]
            if not level_df.empty:
                strict = level_df[level_df['is_ind'] == True]
                approx = level_df[(level_df['confidence'] < 1.0) & (level_df['confidence'] >= self.min_confidence)]
                
                print(f"\nLevel {level}:")
                print(f"  Strict INDs (conf=1.0): {len(strict)}")
                print(f"  Approximate INDs (conf≥{self.min_confidence}): {len(approx)}")
        
        # Confidence distribution
        print(f"\nConfidence Distribution:")
        print("-" * 70)
        
        if not results_df.empty:
            bins = [0, 0.5, 0.8, 0.9, 0.95, 0.99, 1.0]
            labels = ['<0.5', '0.5-0.8', '0.8-0.9', '0.9-0.95', '0.95-0.99', '1.0']
            results_df['conf_bin'] = pd.cut(results_df['confidence'], bins=bins, labels=labels, include_lowest=True)
            
            for label in labels:
                count = len(results_df[results_df['conf_bin'] == label])
                if count > 0:
                    print(f"  {label}: {count} dependencies")
        
        # Find strongest dependencies (confidence=1.0 with largest dep_count)
        strict_inds = results_df[results_df['is_ind'] == True]
        if not strict_inds.empty:
            print(f"\nStrongest Strict Dependencies (sorted by dep_count, top 10):")
            print("-" * 70)
            top_strict = strict_inds.nlargest(10, 'dep_count')
            for _, row in top_strict.iterrows():
                print(f"  {row['dependent']:35s} ⊆ {row['referenced']:35s} "
                      f"dep_count={row['dep_count']:,}, conf={row['confidence']:.6f}")


def run_full_ind_analysis(df: pd.DataFrame, min_confidence: float = 0.95, max_level: int = 2):
    """
    Run complete inclusion dependency analysis (using all data)
    
    Args:
        df: DataFrame (all data)
        min_confidence: Minimum confidence threshold
        max_level: Maximum combination level (1=single column, 2=two columns)
    """
    print("=" * 80)
    print("Full Multi-Column Inclusion Dependency Discovery (with Confidence) - Using All Data")
    print("=" * 80)
    print(f"Data shape: {df.shape[0]:,} rows × {df.shape[1]} columns")
    
    # Data preprocessing: keep only columns with shorter names (avoid combinatorial explosion)
    all_cols = list(df.columns)
    
    # Optional: filter out columns that are not suitable (e.g., long text columns)
    excluded_patterns = ['Violations', 'Address', 'Location', 'DBA Name']  # Exclude long text columns
    cols_to_analyze = [col for col in all_cols if not any(p in col for p in excluded_patterns)]
    
    print(f"Number of columns to analyze: {len(cols_to_analyze)} (excluded long text columns)")
    print(f"Column names: {cols_to_analyze}")
    
    # Use all data, no sampling
    sample_df = df[cols_to_analyze].copy()
    
    # Preprocessing: fill null values and convert to string
    print("\nPreprocessing data...")
    for col in cols_to_analyze:
        sample_df[col] = sample_df[col].fillna('NULL')
        sample_df[col] = sample_df[col].astype(str)
    print("Preprocessing complete")
    
    # Run discovery algorithm
    discoverer = FullINDDiscovery(sample_df, min_confidence=min_confidence, max_level=max_level)
    results = discoverer.discover_all_inds()
    discoverer.print_summary(results)
    
    return results


# ============================================================================
# Execute Analysis
# ============================================================================

if __name__ == "__main__":
    # Check if data exists
    if 'df' not in locals():
        print("Error: Please load data into df variable first")
        print("Example:")
        print("  import pandas as pd")
        print("  df = pd.read_csv(r'E:\\Download\\Food_Inspections_20260327.csv', low_memory=False)")
        print("  # Then run your data cleaning code")
    else:
        # Run complete analysis (using all data)
        ind_results = run_full_ind_analysis(
            df, 
            min_confidence=0.95,  # Only show dependencies with confidence >= 95%
            max_level=2          # Analyze single and two-column dependencies
        )
        
        # Save results to CSV
        if not ind_results.empty:
            output_file = 'inclusion_dependencies_results_full.csv'
            ind_results.to_csv(output_file, index=False)
            print(f"\nResults saved to: {output_file}")
            
            # Display summary statistics
            print("\n" + "=" * 80)
            print("Summary Statistics:")
            print("=" * 80)
            print(f"Total dependencies: {len(ind_results)}")
            print(f"Strict dependencies (conf=1.0): {len(ind_results[ind_results['is_ind'] == True])}")
            print(f"Approximate dependencies (conf<1.0): {len(ind_results[ind_results['is_ind'] == False])}")
            
            # Statistics by level
            print(f"\nStatistics by Level:")
            for level in [1, 2]:
                level_df = ind_results[ind_results['level'] == level]
                print(f"  Level {level}: {len(level_df)} dependencies")
```
